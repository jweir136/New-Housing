## Step 2: Understanding Features

**Sale Price**: What we are trying to predict.  This is the price of the house in dollars.

**MSSubClass**: This is the building class.  A categorical integer with no missing values.

**MSZoning**: This is the general zoning classification.  A categorical String with no missing values.

**LotFrontage**: This is the amount of street that is connected to the property.  This is a float value with some missing data.  This could be due to some homes not being connected to any street.  *This may be solved by setting missing values to zero.*

**LotArea**: The sqft of area.  An int value which is a continous feature.  There is no missing values.

**Street**: The type of road access.  Cat. strings with no missing values.  Most of values are paved.  May not be so important.

**Alley**: The type of alley access.  A cat. string with a ton of missing values.  This is due to a house not having any alley.  *Solve by setting nan values to 'None'*.

**LotShape**: This is the est. shape of the property.  This ranges from 'Reg'=Regular, and 3 other irregular shapes.

**LandContour**: This is the slope of the land.  Lvl=Level property, HLS=a hill, LOW=Low Grade, etc.

**Utilities**: The type of utilities that are available.  This may not be important as only a single home does not have utilities.

**LotConfig**: The configuration of the lot.  This is the location of the home on the street.
